# **Dictionary of Databases**


---


* ABS_T01_TVP = Age and Sex of TVP Residents 
* ABS_T02_TVP = (Legal) Marital Status of TVP Residents 
* ABS_T03_TVP = (Defacto) Marital Status of TVP Residents E.g. Same Sex
* ABS_T04_TVP = Indigenous Status of TVP Residents
* ABS_T08_TVP = English Speaking information of TVP Residents
* ABS_T12_TVP = Highest School Year education of TVP Residents
* ABS_T15_TVP = Volunteer Status of TVP Residents
* ABS_T20_TVP = Family Status of TVP Residents
* ABS_T23_TVP = Household Numbers of TVP Residents
* ABS_T24_TVP = Dwelling Type of TVP Residents
* ABS_T25_TVP = Mortgage Amount of TVP Residents
* ABS_T26_TVP = Rent Amount of TVP Residents
* ABS_T27_TVP = Internet Status of TVP Residents
* ABS_T28_TVP = Number of Bedrooms of TVP Residents
* ABS_T29_TVP = Labour Force, Education and Migration of TVP Residents
* ABS_T32_TVP = Field of Study of TVP Residents
* ABS_T33_TVP = Labour Status of TVP Residents
* ABS_T34_TVP = Occupational Industry of TVP Residents
* ABS_T35_TVP = Occupational Type of TVP Residents

```
merge_year = pd.merge(ABS_TVP_A_2006, ABS_TVP_A_2011, how='outer', on=['Variant Code', 'Variant','Age','Sex','Category','Code Prefix','Code'])
ABS_TVP_A_Merged = pd.merge(merge_year, ABS_TVP_A_2016, how='outer', on=['Variant Code', 'Variant','Age','Sex','Category','Code Prefix','Code'])
ABS_TVP_A_Merged.rename(columns={"Value_x": "2006",'Value_y': "2011", 'Value':"2016"}, inplace = True)
ABS_TVP_A_Merged = ABS_TVP_A_Merged.filter(['Sex','Age','2006','Category','Variant','Code Prefix','Code','2011','2016'], axis=1)
```
