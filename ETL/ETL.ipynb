# **Dictionary of Databases**


---


* ABS_T01_TVP = Age and Sex of TVP Residents 
* ABS_T02_TVP = (Legal) Marital Status of TVP Residents 
* ABS_T03_TVP = (Defacto) Marital Status of TVP Residents E.g. Same Sex
* ABS_T04_TVP = Indigenous Status of TVP Residents
* ABS_T08_TVP = English Speaking information of TVP Residents
* ABS_T12_TVP = Highest School Year education of TVP Residents
* ABS_T15_TVP = Volunteer Status of TVP Residents
* ABS_T20_TVP = Family Status of TVP Residents
* ABS_T23_TVP = Household Numbers of TVP Residents
* ABS_T24_TVP = Dwelling Type of TVP Residents
* ABS_T25_TVP = Mortgage Amount of TVP Residents
* ABS_T26_TVP = Rent Amount of TVP Residents
* ABS_T27_TVP = Internet Status of TVP Residents
* ABS_T28_TVP = Number of Bedrooms of TVP Residents
* ABS_T29_TVP = Labour Force, Education and Migration of TVP Residents
* ABS_T32_TVP = Field of Study of TVP Residents
* ABS_T33_TVP = Labour Status of TVP Residents
* ABS_T34_TVP = Occupational Industry of TVP Residents
* ABS_T35_TVP = Occupational Type of TVP Residents

Drop unnecessary columns and Re index

```
ABS_T01_TVP = ABS_T01_AgeSex_TVP.filter(['SEX_ABS','Sex','Age','Year','Value'], axis=1)
ABS_T01_TVP = ABS_T01_TVP.reset_index(drop=True)

ABS_T02_TVP = ABS_T02_MaritalStatus_TVP.filter(['MSTP_2016','Registered marital status','Age','SEX_ABS','Sex','Year','Value'], axis=1)
ABS_T02_TVP = ABS_T02_TVP.reset_index(drop=True)

ABS_T03_TVP = ABS_T03_SocialMartialStatus_TVP.filter(['MDCP_2016','Social marital status','Age','SEX_ABS','Sex','Year','Value'], axis=1)
ABS_T03_TVP = ABS_T03_TVP.reset_index(drop=True)

ABS_T04_TVP = ABS_T04_Indigenous_TVP.filter(['INGP_2016','Indigenous status','Age','SEX_ABS','Sex','Year','Value'], axis=1)
ABS_T04_TVP = ABS_T04_TVP.reset_index(drop=True)

ABS_T08_EnglishSpeaking_TVP.rename(columns={ABS_T08_EnglishSpeaking_TVP.columns[1]: "English Speaking"}, inplace = True)
ABS_T08_TVP = ABS_T08_EnglishSpeaking_TVP.filter(['ENGLP_2016','English Speaking','Age','SEX_ABS','Sex','Year','Value'], axis=1)
ABS_T08_TVP = ABS_T08_TVP.reset_index(drop=True)

ABS_T12_TVP = ABS_T12_SchoolYear_TVP.filter(['HSCP_2016','Highest Year of School Completed','Age','SEX_ABS','Sex','Census year','Value'], axis=1)
ABS_T12_TVP.rename(columns={"Census year": "Year"}, inplace = True)
ABS_T12_TVP = ABS_T12_TVP.reset_index(drop=True)

ABS_T15_TVP = ABS_T15_Volunteer_TVP.filter(['VOLWP_2016','Voluntary Work for an Organisation or Group','Age','SEX_ABS','Sex','Census year','Value'], axis=1)
ABS_T15_TVP.rename(columns={"Census year": "Year"}, inplace = True)
ABS_T15_TVP = ABS_T15_TVP.reset_index(drop=True)

ABS_T12_TVP = ABS_T20_Family_TVP.filter(['FBLF_2016	','Family Blending','Census year','Value'], axis=1)
ABS_T12_TVP.rename(columns={"Census year": "Year"}, inplace = True)
ABS_T12_TVP = ABS_T12_TVP.reset_index(drop=True)
```
